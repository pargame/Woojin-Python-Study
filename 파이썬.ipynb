## [Chapter 0. 우리가 왜 프로그래밍 언어를 배우는가?]
> 우리는 컴퓨터의 능력을 활용하기 위해 배웁니다.

### 컴퓨터의 능력 2가지
1. 인간보다 훨씬 빨리 정확하게 **계산**합니다.
2. 인간보다 훨씬 많이 정확하게 **기억, 저장**합니다.

### 컴퓨터는 어떻게 그렇게 계산과 저장을 잘할까?
> 컴퓨터는 0과 1로 모든 정보를 저장하고, 계산합니다. 특히, 전기를 이용하기 때문에 속도가 전기만큼 빠르죠.

### 프로그래밍 언어는 무엇인가?
> 일단, 컴퓨터를 활용하기 위해 프로그래밍 언어(aka. 코딩)를 배운다는 것을 알겠죠?
* 0과 1로만 이루어진 컴퓨터에게 *0011100101* 이런 식으로 얘기하긴 인간 입장에서 힘들겠죠?
* 따라서 좀 더 인간 친화적인 명령을 내리기 위해서 개발이 된 *컴퓨터 전용 명령 언어*가 바로 **프로그래밍 언어**입니다.

### 프로그래밍 언어는 어떻게 인간 친화적인 말을 구사할 수 있는 걸까?
> 중간에 **컴파일러** 라는 번역 기계가 존재합니다. 우리가 **프로그래밍 언어**로 작성한 코드를 컴파일러가 *0100110* 이런 식으로 변환시켜주죠.

### 프로그래밍 언어는 하나만 있을까?
> 아니요! 다양한 종류의 언어가 있습니다. 그 언어마다 다른 컴파일러가 필요하겠죠?
* 예를 들어, `C`, `C++`, `C#`, `Python`, `Java`, `JavaScript`, `TypeScript`, `Rust` 등이 있습니다.

### 자, 그러면 이제 Python의 세계로 떠나서 일단 **계산**과 **저장**에 대해 알아볼까요~?

## [Chapter 1. 컴퓨터는 계산과 저장을 어느 대상에게 할까?]
> 바로 **데이터** 입니다. 데이터는 사실은 0과 1로 이루어진 숫자에 불과하지만, 0과 1로 모든 것을 표현할 수가 있습니다.

### 데이터로 모든 것을 어떻게 표현할까요?
> 현실에도 다양한 정보가 있습니다. 숫자 뿐만 아니라, 글자, 그림, 동영상 등이 있습니다.

> 종류가 다양한 현실 세계의 정보를 컴퓨터의 0과 1로만 바꿔야 하겠죠!

> 그렇다면, 방법은 하나밖에 없습니다. 바로, 다양한 정보들을 종류별로 분류하는 것입니다.

> 그러한 종류를 **타입**이라고 부르죠!

### [Chapter 2. 컴퓨터에게 계산은 어떻게 시켜야 하나요?]

## [Chapter 3. 컴퓨터는 데이터를 어떻게 저장할까?]
> 컴퓨터는 데이터를 **변수**에 저장합니다.

### 변수란 무엇일까요?

> 수학에서의 변수와 코딩에서의 변수는 엄연히 다릅니다.

> 코딩에서의 변수는 *변하는 수*가 아니고, **공동 보관함**처럼 생각해야 합니다.

> 따라서 우리는 코딩을 통해 *공동 보관함*에 데이터를 보관하며 저장이 이루어지는 것입니다.

### 변수, 즉 공동 보관함은 어떻게 만들고, 저장을 해야 하나요?

> 그러기 위해서는 우리는 `=` 이라는 기호를 완벽하게 알고 사용할 줄 알아야 합니다.

### `=` 이라는 '기호'요? 그렇다면 연산자가 아닌건가요?

> 연산자로 분류가 되긴 하지만, 명백하게 다른 점이 있습니다. 바로, **반환값**이 없다는 점이죠.

> 자, 그럼 이제 변수에 저장하는 기호인 `=`에 대해 자세히 알아봅시다.

### 변수와 `=` 사용 형식과 그 예시

|형식|예시1|예시2|잘못된 예시1|잘못된 예시2|
|:---:|:---:|:---:|:---:|:---:|
|`변수` `=` `데이터`|`a=3`|`b='ab'`|`3=n`|`'ab'`=`'c'`|

> 주의할 점은, `=` 왼쪽에는 항상 **변수**가 오는 변수 자리, 오른쪽은 항상 **데이터**가 오는 데이터 자리입니다.

> 오른쪽의 **데이터**를 왼쪽의 **변수**에 저장하는 것입니다.

### 그럼 내가 공동 보관함에 언제든지 원하는 데이터를 '저장'하는 것이 가능하겠네요? 근데.. 저장한 데이터를 꺼내보려면요?

> 여기서 *변수의 난해함*이 등장합니다. 변수는 사실 꺼내보는 행위가 따로 존재하지 않습니다.

> **변수를 데이터 자리에 넣어서 코딩하면, 그냥 그 변수에 저장된 데이터로 처리가 됩니다.**

> 즉, 데이터 자리의 변수는 데이터, 변수 자리의 변수는 변수로 취급되는 *이중적인* 것이 바로 **변수**입니다.

### 그러면 변수 자리 데이터 자리 어떻게 구분해요? 너무 어려워요ㅠ

> 간단합니다. 변수 자리는 딱 하나입니다. 바로 **`=` 왼쪽** 이죠. 쉽죠?ㅎ

> *참고로 Chpater 7에서 변수의 다른 자리도 있다는 것을 배웁니다.*

## [Chapter 4. 인덱싱]

> 챕터 2 **연산자**를 배울 때, 인덱싱을 봤었죠?

> 우리가 타입을 배우면서, str과 list를 이미 알고 있습니다.

> 이제는, 좀 더 깊이 알아봅시다.

### int와 str의 명백하게 다른 점을 알아봅시다.

> 만약에 다음과 같은 코드가 있다고 생각해봅시다.

`A=3`

`B='abc'`

> 이 때, 데이터 자리에 오는 변수는 데이터로 처리가 된다고 했죠?

> 그렇다면 데이터 자리에 오는 `A`와 `B`가 각각 어떤 의미로 처리될까요?

> `A`는 쉽죠! 그냥 `3` 이라는 int 타입의 데이터로 처리됩니다.

> `B`는 마찬가지로 `'abc'`라는 str 타입의 데이터로 처리됩니다.

> 자, 여기서 `A`와 `B`는 각각 의미하는 것이 다릅니다. *Why?*

> `A`는 하나의 데이터지만, `B`는 `'a'`, `'b'`, `'c'` 3개의 데이터를 한꺼번에 의미하고 있습니다.

### 그렇다면 여기서 만약에 개발자가 `B` 변수 안에 있는 `'abc'` 중에서 `'b'`라는 데이터를 하나만 가져와서 처리하고 싶으면요??

> 그래서 바로 **인덱싱**을 알아야 합니다.

> 정리하자면, 여러 개의 데이터를 의미하는 타입은 누가 있죠? 바로 str/list가 있습니다. *(추후에 더 많은 타입을 배울 예정)*

> 이렇게 여러 개를 보유하고 있는 타입은 전문 용어로 **Iterable** 타입이라고 부릅니다.

> 'Iterable'을 직역하자면, '셀 수 있는' 이라는 뜻이죠.

> 그러면 이제 **Iterable** 타입인 str/list 에서 인덱싱 하는 법을 알아볼까요?

### 인덱싱 하는 법

> 먼저, 인덱싱이란 순수 데이터가 아니라 '변수'에 적용하는 연산자입니다.

> 따라서 다음과 같은 변수들이 먼저 만들어져 있다고 가정해봅시다.

`A=3`

`B='ab12'`

`C=[1,2,[],'ab']`

> A는 int 타입으로 처리가 될 것이고, B와 C도 각각 str, list로 처리가 되겠죠?

### 인덱싱 형식과 예시

|형식|예시1|예시1의 반환값|예시2|예시2의 반환값|예시3|예시3의 반환값|예시4|예시4의 반환값|
|:---:|:---:|:---:|:---:|:---:|:---:|:---:|:---:|:---:|
|`변수` `[` `인덱스` `]`|`B[2]`|`'1'`|`C[0]`|`1`|`B[0]`|`'a'`|`A[0]`|`error!`|

### 어? 왜 B의 2 인덱스가 `'b'`가 아니고 `'1'`이 나오나요?

> 왜냐하면 **인덱스는 1이 아니라 0부터 시작하기 때문입니다.**

### `A[0]`은 왜 에러가 뜰까요?

> 왜냐하면 `A`는 `Iterable` 타입이 아니기 때문입니다. **인덱싱은 무조건 `Iterable` 타입에만 적용이 가능합니다.**

> 참고로, Iterable 타입에서 들어있는 하나의 값을 **'원소'** 혹은 **'요소'** 라고 합니다.

> 즉 위에서 `B[2]`의 반환값을 말로 풀어 쓰면, *B의 2 인덱스의 원소*가 되겠죠?
