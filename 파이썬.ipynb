## [Chapter 0. 우리가 왜 프로그래밍 언어를 배우는가?]
> 우리는 컴퓨터의 능력을 활용하기 위해 배웁니다.

### 컴퓨터의 능력 2가지
1. 인간보다 훨씬 빨리 정확하게 **계산**합니다.
2. 인간보다 훨씬 많이 정확하게 **기억, 저장**합니다.

### 컴퓨터는 어떻게 그렇게 계산과 저장을 잘할까?
> 컴퓨터는 0과 1로 모든 정보를 저장하고, 계산합니다. 특히, 전기를 이용하기 때문에 속도가 전기만큼 빠르죠.

### 프로그래밍 언어는 무엇인가?
> 일단, 컴퓨터를 활용하기 위해 프로그래밍 언어(aka. 코딩)를 배운다는 것을 알겠죠?
* 0과 1로만 이루어진 컴퓨터에게 *0011100101* 이런 식으로 얘기하긴 인간 입장에서 힘들겠죠?
* 따라서 좀 더 인간 친화적인 명령을 내리기 위해서 개발이 된 *컴퓨터 전용 명령 언어*가 바로 **프로그래밍 언어**입니다.

### 프로그래밍 언어는 어떻게 인간 친화적인 말을 구사할 수 있는 걸까?
> 중간에 **컴파일러** 라는 번역 기계가 존재합니다. 우리가 **프로그래밍 언어**로 작성한 코드를 컴파일러가 *0100110* 이런 식으로 변환시켜주죠.

### 프로그래밍 언어는 하나만 있을까?
> 아니요! 다양한 종류의 언어가 있습니다. 그 언어마다 다른 컴파일러가 필요하겠죠?
* 예를 들어, `C`, `C++`, `C#`, `Python`, `Java`, `JavaScript`, `TypeScript`, `Rust` 등이 있습니다.

### 자, 그러면 이제 Python의 세계로 떠나서 일단 **계산**과 **저장**에 대해 알아볼까요~?

## [Chapter 1. 컴퓨터는 계산과 저장을 어느 대상에게 할까?]
> 바로 **데이터** 입니다. 데이터는 사실은 0과 1로 이루어진 숫자에 불과하지만, 0과 1로 모든 것을 표현할 수가 있습니다.

### 데이터로 모든 것을 어떻게 표현할까요?
> 현실에도 다양한 정보가 있습니다. 숫자 뿐만 아니라, 글자, 그림, 동영상 등이 있습니다.

> 종류가 다양한 현실 세계의 정보를 컴퓨터의 0과 1로만 바꿔야 하겠죠!

> 그렇다면, 방법은 하나밖에 없습니다. 바로, 다양한 정보들을 종류별로 분류하는 것입니다.

> 그러한 종류를 **타입**이라고 부르죠!

### 데이터의 타입은 어떤 것들이 있나요?
> 일단, 프로그래밍 언어마다 타입이 다 다르게 존재합니다. 여기선 *Python*의 타입에 대해 다룹시다.

|별칭|이름|설명|예시1|예시2|예시3|예시4|주의할점|
|:---:|:---:|:---:|:---:|:---:|:---:|:---:|:---:|
|간단수|int|점이 없는 숫자|`3`|`0`|`-3`|`123`|X|
|복잡수|float|소수점이 있는 숫자|`3.14`|`2.3`|`-3.6543`|`3.0`|`3.0` 같은 경우, 수학에서는 '정수형'으로 분류가 됩니다. 하지만 컴퓨터 세계에선, 소수점이 존재하기 때문에 int와 전혀 다른 float타입이 됩니다.|
|참거짓|bool|참을 의미하거나 거짓을 의미하는 데이터|`True`|`False`|X|X|대문자로 시작!|

|포장품|list|대괄호로 여러 데이터를 포장한 데이터|`[1,2,3]`|`['ab','cd']`|`[]`|`[True, [1,'a'],None,'st12']`|들어간 데이터 사이마다 콤마를 넣어주기|

> 이제, 어떤 데이터를 보면 그 데이터의 타입을 반드시 말할 수 있겠죠?

> 다음으로, **계산**과 **저장**을 시키는 법을 배우러 떠나봅시다!

### [Chapter 2. 컴퓨터에게 계산은 어떻게 시켜야 하나요?]
> 계산은 정해진 처리 순서에 따라 **연산자**를 기준으로 처리가 됩니다.

> 계산의 결과를 **반환값**이라고 하죠. 반환값은 임시적으로 결과로써 나오는 데이터입니다.

### 그러면 연산자가 어떤 것들이 있나요?
> 크게 보면 **산술**, **비교**, **논리**, **인덱싱**이 있습니다. 이것들을 알아보러 가볼까요?

## [Chapter 3. 컴퓨터는 데이터를 어떻게 저장할까?]
> 컴퓨터는 데이터를 **변수**에 저장합니다.

### 변수란 무엇일까요?

> 수학에서의 변수와 코딩에서의 변수는 엄연히 다릅니다.

> 코딩에서의 변수는 *변하는 수*가 아니고, **공동 보관함**처럼 생각해야 합니다.

> 따라서 우리는 코딩을 통해 *공동 보관함*에 데이터를 보관하며 저장이 이루어지는 것입니다.

### 변수, 즉 공동 보관함은 어떻게 만들고, 저장을 해야 하나요?

> 그러기 위해서는 우리는 `=` 이라는 기호를 완벽하게 알고 사용할 줄 알아야 합니다.
