## [Chapter 0. 우리가 왜 프로그래밍 언어를 배우는가?]
> 우리는 컴퓨터의 능력을 활용하기 위해 배웁니다.

### 컴퓨터의 능력 2가지
1. 인간보다 훨씬 빨리 정확하게 **계산**합니다.
2. 인간보다 훨씬 많이 정확하게 **기억, 저장**합니다.

### 컴퓨터는 어떻게 그렇게 계산과 저장을 잘할까?
> 컴퓨터는 0과 1로 모든 정보를 저장하고, 계산합니다. 특히, 전기를 이용하기 때문에 속도가 전기만큼 빠르죠.

### 프로그래밍 언어는 무엇인가?
> 일단, 컴퓨터를 활용하기 위해 프로그래밍 언어(aka. 코딩)를 배운다는 것을 알겠죠?
* 0과 1로만 이루어진 컴퓨터에게 *0011100101* 이런 식으로 얘기하긴 인간 입장에서 힘들겠죠?
* 따라서 좀 더 인간 친화적인 명령을 내리기 위해서 개발이 된 *컴퓨터 전용 명령 언어*가 바로 **프로그래밍 언어**입니다.

### 프로그래밍 언어는 어떻게 인간 친화적인 말을 구사할 수 있는 걸까?
> 중간에 **컴파일러** 라는 번역 기계가 존재합니다. 우리가 **프로그래밍 언어**로 작성한 코드를 컴파일러가 *0100110* 이런 식으로 변환시켜주죠.

### 프로그래밍 언어는 하나만 있을까?
> 아니요! 다양한 종류의 언어가 있습니다. 그 언어마다 다른 컴파일러가 필요하겠죠?
* 예를 들어, `C`, `C++`, `C#`, `Python`, `Java`, `JavaScript`, `TypeScript`, `Rust` 등이 있습니다.

### 자, 그러면 이제 Python의 세계로 떠나서 일단 **계산**과 **저장**에 대해 알아볼까요~?

## [Chapter 1. 컴퓨터는 계산과 저장을 어느 대상에게 할까?]
> 바로 **데이터** 입니다. 데이터는 사실은 0과 1로 이루어진 숫자에 불과하지만, 0과 1로 모든 것을 표현할 수가 있습니다.

### 데이터로 모든 것을 어떻게 표현할까요?
> 현실에도 다양한 정보가 있습니다. 숫자 뿐만 아니라, 글자, 그림, 동영상 등이 있습니다.

> 종류가 다양한 현실 세계의 정보를 컴퓨터의 0과 1로만 바꿔야 하겠죠!

> 그렇다면, 방법은 하나밖에 없습니다. 바로, 다양한 정보들을 종류별로 분류하는 것입니다.

> 그러한 종류를 **타입**이라고 부르죠!

## [Chapter 3. 컴퓨터는 데이터를 어떻게 저장할까?]
> 컴퓨터는 데이터를 **변수**에 저장합니다.

### 변수란 무엇일까요?

> 수학에서의 변수와 코딩에서의 변수는 엄연히 다릅니다.

> 코딩에서의 변수는 *변하는 수*가 아니고, **공동 보관함**처럼 생각해야 합니다.

> 따라서 우리는 코딩을 통해 *공동 보관함*에 데이터를 보관하며 저장이 이루어지는 것입니다.

### 변수, 즉 공동 보관함은 어떻게 만들고, 저장을 해야 하나요?

> 그러기 위해서는 우리는 `=` 이라는 기호를 완벽하게 알고 사용할 줄 알아야 합니다.

## `=` 이라는 '기호'요? 그렇다면 연산자가 아닌건가요?

> 연산자로 분류가 되긴 하지만, 명백하게 다른 점이 있습니다. 바로, **반환값**이 없다는 점이죠.

> 자, 그럼 이제 변수에 저장하는 기호인 `=`에 대해 자세히 알아봅시다.

|형식|예시1|예시2|잘못된 예시1|잘못된 예시2|
|:---:|:---:|:---:|:---:|:---:|
|`변수` `=` `데이터`|`a=3`|`b='ab'`|`3=n`|`'ab'`=`'c'`|

> 주의할 점은, `=` 왼쪽에는 항상 **변수**가 오는 변수 자리, 오른쪽은 항상 **데이터**가 오는 데이터 자리입니다.

> 오른쪽의 **데이터**를 왼쪽의 **변수**에 저장하는 것입니다.

### 그럼 내가 공동 보관함에 언제든지 원하는 데이터를 '저장'하는 것이 가능하겠네요? 근데.. 저장한 데이터를 꺼내보려면요?

> 여기서 *변수의 난해함*이 등장합니다. 변수는 사실 꺼내보는 행위가 따로 존재하지 않습니다.

> **변수를 데이터 자리에 넣어서 코딩하면, 그냥 그 변수에 저장된 데이터로 처리가 됩니다.**

> 즉, 데이터 자리의 변수는 데이터, 변수 자리의 변수는 변수로 취급되는 *이중적인* 것이 바로 **변수**입니다.

### 그러면 변수 자리 데이터 자리 어떻게 구분해요? 너무 어려워요ㅠ

> 간단합니다. 변수 자리는 딱 하나입니다. 바로 **`=` 왼쪽** 이죠. 쉽죠?ㅎ

> *참고로 Chpater 7에서 변수의 다른 자리도 있다는 것을 배웁니다.*

## [Chapter 4. 인덱싱]

> 인덱싱이란 것은 챕터2에서 **연산자**에서 처음 확인했습니다.

> 따라서 인덱싱은 `+`처럼 **연산자의 일종**이죠!

> 만약에 다음과 같은 코드가 있다고 생각해봅시다.

`A=3`

`B='abc'`

> 이 때, 데이터 자리에 오는 변수는 데이터로 처리가 된다고 했죠?

> 자, 여기서 `A`와 `B`는 각각 의미하는 것이 다릅니다. *Why?*

> `A`는 하나의 데이터지만, `B`는 `'a'`, `'b'`, `'c'` 3개의 데이터를 한꺼번에 의미하고 있습니다.

### 그렇다면 여기서 만약에 개발자가 `B` 변수 안에 있는 `'abc'` 중에서 `'b'`라는 데이터를 하나만 가져와서 처리하고 싶으면요??

> 그래서 바로 **인덱싱**을 알아야 합니다.

> 여러 개의 데이터를 의미하는 타입은 누가 있죠? 바로 str/list가 있습니다. *(추후에 더 많은 타입을 배울 예정)*

> 이렇게 여러 개를 보유하고 있는 타입은 전문 용어로 **Iterable** 타입이라고 부릅니다.

> 'Iterable'을 직역하자면, '셀 수 있는' 이라는 뜻이죠.

> 그러면 이제 **Iterable** 타입인 str/list 에서만 인덱싱이 가능하겠죠? 인덱싱 하는 법을 알아봅시다.

### 인덱싱 하는 법

> 먼저, 인덱싱이란 순수 데이터가 아니라 '변수'에 적용하는 연산자입니다.

> 따라서 다음과 같은 변수들이 먼저 만들어져 있다고 가정해봅시다.

`A=3`

`B='ab12'`

`C=[1,2,[],'ab']`

> A는 int 타입으로 처리가 될 것이고, B와 C도 각각 str, list로 처리가 되겠죠?

|형식|예시1|예시1의 반환값|예시2|예시2의 반환값|예시3|예시3의 반환값|예시4|예시4의 반환값|
|:---:|:---:|:---:|:---:|:---:|:---:|:---:|:---:|:---:|
|`변수` `[` `인덱스` `]`|`B[2]`|`'1'`|`C[0]`|`1`|`B[0]`|`'a'`|`A[0]`|`error!`|

### 어? 왜 B의 2 인덱스가 `'b'`가 아니고 `'1'`이 나오나요?

> 왜냐하면 **인덱스는 1이 아니라 0부터 시작하기 때문입니다.**

### `A[0]`은 왜 에러가 뜰까요?

> 왜냐하면 `A`는 `Iterable` 타입이 아니기 때문입니다. **인덱싱은 무조건 `Iterable` 타입에만 적용이 가능합니다.**

> 참고로, Iterable 타입에서 들어있는 하나의 값을 **'원소'** 혹은 **'요소'** 라고 합니다.

> 즉 위에서 `B[2]`의 반환값을 말로 풀어 쓰면, *B의 2 인덱스의 원소*가 되겠죠?

### 정리하기

> str, list 처럼 여러개의 원소를 보유하는 타입을 *Iterable* 타입이라고 한다.

> 여러개 중에서 하나를 꺼내기 위해 **인덱싱**을 사용해야 한다.

> 인덱싱은 `변수[인덱스]` 형식으로 사용한다.

> 인덱스란, 해당 원소의 순서로, **0부터 시작한다**

## [Chapter 5. if 구문]

> 다음과 같은 상황을 생각해보자.

`A=3412321`

`B=1234242`

> 두 줄의 코드가 있다. `A*B`의 값을 알고 싶다.

> 단순하게 `*` 연산자를 사용하면 된다.

> 반면에, `A*B`가 짝수가 나오는지 홀수가 나오는지 알고 싶다.

> 우리가 배운 `%`를 활용하면 된다.

`A*B%2==0`

> 모든 연산자의 일이 끝나면 반환값으로 `True`나 `False`가 나올 것이다.

> `True`가 나오면 짝수일테고, `False`가 나오면 홀수일테니까.

> 그러면, `A*B`가 짝수면 "짝수입니다."를 출력하고 싶다면?

> `A*B`가 홀수면 "홀수입네다."를 출력하고 싶다면?

> 이 때 문제가 발생한다. 짝수일 때랑 홀수일 때랑 출력해야할 것이 다르기 때문이다...!

### 만약에 `A*B`가 짝수면 어떤 코드가 실행되어야 하나요?

`print("짝수입니다")`

### 만약에 홀수면요?

`print("홀수입네다")`

### 그러면 `A*B` 결과의 따라서 실행될 코드가 달라지네요?

> 그렇다! 그렇기 때문에, **if 구문**을 잘 쓸 줄 알아야 한다.

# if 구문 사용법1: **if**

> `if` 라는 키워드는 단순하게 **조건**이 `True`면 실행하고, `False`면 실행하지 않는다.

    >>예시
    if True:
        print('hi')

> 여기서 **조건** 은 `True`기 때문에 `print('hi')`가 실행된다!

> 그러면 위에서 봤던 상황은 해결될 수 있다!

    >>if문으로 해결!!
    if A*B%2==0:
        print('짝수입니다')
    if A*B%2==1:
        print('홀수입네다')

# if 구문 사용법2: **else**

> `else`라는 키워드는 **혼자 쓸 수 없다.**

> 반드시 `if` 키워드 아래에 옵션처럼 붙어야 한다.

> 또한 **조건도 없다.**

> 그럼 무슨 뜻으로 사용할까?

> 이렇게 생각하고 사용해야 한다.

**"else: 만나면 실행!"**

> 왜 이렇게 생각해야 하는지 알아보자.

> 일단 위의 문제를 if-else 로 짜보겠다.

    >>예시
    if A*B%2==0:-------------------(1)
        print("짝수입니다")--------(2)
    else:--------------------------(3)
        print("홀수입네다")--------(4)
    (5: 그 다음 코드..)
    
> 만약에 `A*B` 짝수라면 어떻게 실행될까?

(1)->(2)->(5)

> 왜 (3)은 보지도 않고 (5)으로 갈까?

> 현재 if-else문을 사용하는데, 하나의 if문을 사용중이다

    "핵심 규칙!! 하나의 if문에서는 최대 한 번 실행이 된다."

> 따라서, (1)이 `True`가 나와서 (2)이 실행이 되었으므로, (3)과 (4)은 볼 필요 없다!

> 만약에 `A*B`가 홀수라면 어떻게 실행될까?

(1)->(3)->(4)->(5)

> (1)을 컴퓨터가 보고 나서, 조건이 `False`가 된다는 것을 확인했다.

> 그 다음으로 (2) 당연히 실행되면 안되므로 pass

> (3)을 컴퓨터가 봤다. 아까 else를 어떻게 생각하라고 했는지 기억나는가?!

**"else: 만나면 실행"**

> 컴퓨터가 (3)을 만나지 않았는가? 그러므로 (4)이 실행된다.

> 실행된 후, **if문은 최대 하나만 실행된다** 는 사실을 기억하는가?

> 따라서 if문을 빠져나오고 (5)으로 넘어가는 것이다.

# if 구문 사용법3: **elif**

> elif는 else랑 마찬가지로 if에 붙는 옵션이다.

> 단, if 밑에 붙으면서 else보다는 위에 붙어야 한다.

> 또한 elif는 여러개를 붙일 수 있다.

(새로운 문제 상황)

`A=?`

> 이 때, A가 0보다 크면 "A>0"를 출력하고 싶다.

> A가 0보다 작으면 "A<0"를 출력하고 싶다.

> A가 0과 같으면 "A==0"를 출력하고 싶다.

> 아까 짝수 홀수와 달리, 이번엔 세 가지 경우로 나누어진다.

    >>if-else 풀이
    if A>0:-----------------(1)
        print('A>0')--------(2)
    else:-------------------(3)
        if A<0:-------------(4)
            print('A<0')----(5)
        else:---------------(6)
            print('A==0')---(7)
        (8: 안쪽의 if문 종료)
    (9: 바깥쪽 if문 종료)

[Case 1]
> 만약 A에 5가 저장되어 있다고 해보자.

> 그럼 어떤 번호 순으로 실행될까?

(1)->(2)->(9)

> why? (1)에서 `True`가 조건이 되므로, (2)을 실행한다.

> 그 다음, **하나의 if문에서 하나가 실행이 되었으므로, (3)을 pass하고 (9)으로 간다.**

> 하나의 if문에서 최대 한 개만 실행된다는 점 반드시 기억할 것!!

[Case 2]
> 만약 A에 -3이 저장되어 있다고 해보자.

> 어떤 번호 순으로 실행될까?

(1)->(3)->(4)->(5)->(9)

> 일단, (1)에 갔더니 조건이 `False`가 된다.

> 따라서 (2)은 pass하고 (3)에 간다.

> (3)은 else, 만나면 실행! 따라서 (4)으로 간다.

> 또 if문이 있네? (4)의 if문은 (1)의 if와 또 다른 if문이다.

> 아무튼 (4)의 조건은 `True`! (5)으로 간다.

> (5)이 실행되고, (4)의 if문에서 실행이 하나 되었으므로 (6)을 안가고 (8)으로 간다.

> (8)은 뭐 없죠? 그냥 끝나고, (3)의 else문도 끝!

> 마지막으로, 맨 처음 (1) if문에서도 하나가 실행되었죠!? 따라서 종료되는 (9)으로 갑니다.

    >>if-else 풀이
    if A>0:-----------------(1)
        print('A>0')--------(2)
    else:-------------------(3)
        if A<0:-------------(4)
            print('A<0')----(5)
        else:---------------(6)
            print('A==0')---(7)
        (8: 안쪽의 if문 종료)
    (9: 바깥쪽 if문 종료)

[Case 3]
> 만약 A에 0이 저장되어 있다면? 실행되는 번호를 순서대로 직접 써봅시다.

(1)->(3)->(4)->(6)->(7)->(8)->(9)

# 마지막!! elif를 옵션으로 사용해보기

    (if-elif 풀이)
    if A>0:-------------(1)
        print('A>0')----(2)
    elif A<0:-----------(3)
        print('A<0')----(4)
    elif A==0:----------(5)
        print('A==0')---(6)
    (7: if문 종료)

[Case 4]
> 바로 해봅시다! A에 3이 저장되어 있다면?

(1)->(2)->(7)

반드시 실행이 (2)에서 됐으니까, 나머지 elif는 볼 필요가 없음을 기억하자!

> A에 0이 저장되어 있다면?

(1)->(3)->(5)->(6)->(7)

> A에 -4가 저장되어 있다면? 어떤 순서로 실행될까?

(1)->(3)->(4)->(7)
